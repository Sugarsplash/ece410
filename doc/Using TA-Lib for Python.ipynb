## Intro

The purpose of this notebook is to simply use [pandas_datareader](http://pandas.pydata.org/pandas-docs/stable/remote_data.html) to access stock data then use that data to do technical analysis using [TA-Lib for python](https://mrjbq7.github.io/ta-lib/index.html).

```python
import pandas_datareader.data as web
import datetime
import matplotlib.pylab as plt
%matplotlib inline
from IPython.display import set_matplotlib_formats
import talib as ta
import numpy as np

# start and end dates
start = datetime.datetime(2015, 1, 1)
end = datetime.datetime(2016, 5, 18)

# pull SPY data from yahoo finance using above dates for start and end
f = web.DataReader("SPY", 'yahoo', start, end)

# Graphing helper function
def setup_graph(title='',x_label='', y_label='', fig_size=None):
    fig = plt.figure()
    if fig_size != None:
        fig.set_size_inches(fig_size[0],
                            fig_size[1])
    ax = fig.add_subplot(111)
    ax.set_title(title)
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)

# array to hold closing values
close = []

# there's probably better ways to do this
for element in f['Close']:
    close.append(float(element))
close = np.array(close)

six_sma = ta.SMA(close, timeperiod=6)
six5_sma = ta.SMA(close, timeperiod=65)

setup_graph('Simple Moving Avg', 'Day', 'Moving Avg/Value', (9,6.5))
plt.plot(close)
plt.plot(six_sma)
plt.plot(six5_sma)

plt.legend(['Close', '6-Day SMA', '65-Day SMA'], loc='best')
plt.show()
```

```python
I1 = ta.RSI(close, timeperiod=14)
setup_graph('Relative Strength Index', 'Day', 'RSI', (9,6.5))
plt.plot(I1)
plt.show()
```

```python
#from talib import MA_Type
upper, mid, lower = ta.BBANDS(close, timeperiod=9, nbdevup=2, nbdevdn=2)

setup_graph('Bollinger Bands', 'Day', 'Value', (9,6.5))
plt.plot(upper)
plt.plot(mid)
plt.plot(lower)
plt.legend(['upper', 'middle', 'lower'], loc='best')
plt.show()
```

```python
I2 = (close - upper)/upper
I3 = (close - lower)/lower

setup_graph('Avg distance of bands from actual value', 'Day', 'Distance', (9,6.5))
plt.plot(I2)
plt.plot(I3)
plt.legend(['Upper', 'Lower'], loc='best')
plt.show()
```

```python
high = []
low = []

# there's probably better ways to do this
for element in f['High']:
    high.append(float(element))
high = np.array(high)
for element in f['Low']:
    low.append(float(element))
low = np.array(low)


fastk, fastd = ta.STOCHF(high, low, close, fastk_period=14, fastd_period=3)
I4 = fastk
I5 = fastd

setup_graph('Stochastic Fast', '', '', (9,6.5))
plt.plot(fastk)
plt.plot(fastd)
plt.legend(['fastk', 'fastd'])
plt.show()
```
